# Secci√≥n 1:

Durante el proceso de limpieza de datos se tomaron decisiones diferenciadas seg√∫n la importancia de cada variable. Por un lado, se eliminaron las observaciones que presentaban valores faltantes en la variable price, ya que esta representa la variable principal del an√°lisis y es esencial para futuros modelos de regresi√≥n; imputar valores en este caso podr√≠a introducir sesgos y afectar la validez de los resultados. Por otro lado, se realiz√≥ imputaci√≥n en la variable reviews_per_month utilizando la mediana, debido a que es una caracter√≠stica auxiliar y su ausencia no justifica la eliminaci√≥n completa de registros. El uso de la mediana permite conservar la mayor cantidad de datos posibles y reduce el impacto de valores extremos presentes en el dataset. Estas decisiones se basaron en criterios estad√≠sticos y en la relevancia anal√≠tica de cada variable dentro del estudio.

```python
# 1. Revisi√≥n de los datos
df.info()
df.shape
df.dtypes

# Valores faltantes (Top 20)
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)

# Duplicados
df.duplicated().sum()

# Eliminar duplicados (y verificar)
df = df.drop_duplicates()
df.duplicated().sum()


# 2. Correcci√≥n de fechas
df["last_scraped"] = pd.to_datetime(df["last_scraped"], errors="coerce")
df["host_since"]   = pd.to_datetime(df["host_since"], errors="coerce")

# Verificar tipos y NaT
df[["last_scraped", "host_since"]].dtypes
df[["last_scraped", "host_since"]].isna().sum()


# 3. Conversi√≥n de num√©ricas que est√°n como texto

# Porcentajes -> float (ej. "95%" -> 95.0)
for col in ["host_response_rate", "host_acceptance_rate"]:
    if col in df.columns:
        df[col] = (df[col]
                   .astype(str)
                   .str.replace("%", "", regex=False)
                   .replace("nan", np.nan))
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Booleano t/f -> 1/0
if "host_is_superhost" in df.columns:
    df["host_is_superhost"] = (
    df["host_is_superhost"]
    .astype(str)
    .str.lower()
    .map({"t":1, "true":1, "f":0, "false":0})
)

# Price: convertir si viene como texto con $ o comas
if "price" in df.columns and df["price"].dtype == "object":
    df["price"] = (df["price"].astype(str)
                   .str.replace("$", "", regex=False)
                   .str.replace(",", "", regex=False)
                   .replace("nan", np.nan))
    df["price"] = pd.to_numeric(df["price"], errors="coerce")

# (Opcional: evidencia de que cambi√≥)
df[["host_response_rate", "host_acceptance_rate", "host_is_superhost", "price"]].dtypes


# 4. Manejo de valores faltantes

# Eliminar filas si falta variable clave (price)
df = df.dropna(subset=["price"])

# Imputar reviews_per_month
if "reviews_per_month" in df.columns:
    df["reviews_per_month"] = pd.to_numeric(df["reviews_per_month"], errors="coerce")
    df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].median())



# 5. Verificar cambios
df.info()
df.isnull().sum().sort_values(ascending=False).head(20)
df.describe()
df.shape
```
