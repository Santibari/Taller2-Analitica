# Secci√≥n 1:

Durante el proceso de limpieza de datos se tomaron decisiones diferenciadas seg√∫n la importancia de cada variable. Por un lado, se eliminaron las observaciones que presentaban valores faltantes en la variable price, ya que esta representa la variable principal del an√°lisis y es esencial para futuros modelos de regresi√≥n; imputar valores en este caso podr√≠a introducir sesgos y afectar la validez de los resultados. Por otro lado, se realiz√≥ imputaci√≥n en la variable reviews_per_month utilizando la mediana, debido a que es una caracter√≠stica auxiliar y su ausencia no justifica la eliminaci√≥n completa de registros. El uso de la mediana permite conservar la mayor cantidad de datos posibles y reduce el impacto de valores extremos presentes en el dataset. Estas decisiones se basaron en criterios estad√≠sticos y en la relevancia anal√≠tica de cada variable dentro del estudio.

```python
# 1. Revisi√≥n de los datos
df.info()
df.shape
df.dtypes

# Valores faltantes (Top 20)
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)

# Duplicados
df.duplicated().sum()

# Eliminar duplicados (y verificar)
df = df.drop_duplicates()
df.duplicated().sum()


# 2. Correcci√≥n de fechas
df["last_scraped"] = pd.to_datetime(df["last_scraped"], errors="coerce")
df["host_since"]   = pd.to_datetime(df["host_since"], errors="coerce")

# Verificar tipos y NaT
df[["last_scraped", "host_since"]].dtypes
df[["last_scraped", "host_since"]].isna().sum()


# 3. Conversi√≥n de num√©ricas que est√°n como texto

# Porcentajes -> float (ej. "95%" -> 95.0)
for col in ["host_response_rate", "host_acceptance_rate"]:
    if col in df.columns:
        df[col] = (df[col]
                   .astype(str)
                   .str.replace("%", "", regex=False)
                   .replace("nan", np.nan))
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Booleano t/f -> 1/0
if "host_is_superhost" in df.columns:
    df["host_is_superhost"] = (
    df["host_is_superhost"]
    .astype(str)
    .str.lower()
    .map({"t":1, "true":1, "f":0, "false":0})
)

# Price: convertir si viene como texto con $ o comas
if "price" in df.columns and (df["price"].dtype == "object" or df["price"].dtype == "str"):
    df["price"] = (df["price"].astype(str)
                   .str.replace("$", "", regex=False)
                   .str.replace(",", "", regex=False)
                   .replace("nan", np.nan))
    df["price"] = pd.to_numeric(df["price"], errors="coerce")

# (Opcional: evidencia de que cambi√≥)
df[["host_response_rate", "host_acceptance_rate", "host_is_superhost", "price"]].dtypes


# 4. Manejo de valores faltantes

# Eliminar filas si falta variable clave (price)
df = df.dropna(subset=["price"])

# Imputar reviews_per_month
if "reviews_per_month" in df.columns:
    df["reviews_per_month"] = pd.to_numeric(df["reviews_per_month"], errors="coerce")
    df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].median())



# 5. Verificar cambios
df.info()
df.isnull().sum().sort_values(ascending=False).head(20)
df.describe()
df.shape
```

```python
# ===== AN√ÅLISIS COMPLEMENTARIO: COLINEALIDAD Y OUTLIERS =====

print("\n" + "="*80)
print("AN√ÅLISIS T√âCNICO COMPLEMENTARIO")
print("="*80)

# 1. Matriz de correlaci√≥n entre variables num√©ricas clave
print("\n1. MATRIZ DE CORRELACI√ìN (para detectar colinealidad):")
print("-" * 80)
numeric_cols = ['price', 'bedrooms', 'beds', 'accommodates', 'number_of_reviews']
numeric_cols = [col for col in numeric_cols if col in df.columns]
corr_matrix = df[numeric_cols].corr()
print(corr_matrix.round(3))

# mapa de calor de correlaciones
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor - Correlaciones')
plt.show()

print("\n‚ö†Ô∏è  AN√ÅLISIS DE COLINEALIDAD:")
print(f"   ‚Ä¢ Correlaci√≥n bedrooms-accommodates: {corr_matrix.loc['bedrooms', 'accommodates']:.3f}")
if corr_matrix.loc['bedrooms', 'accommodates'] > 0.8:
    print("     ‚Üí ALTA COLINEALIDAD: Usar solo accommodates en el modelo")
elif corr_matrix.loc['bedrooms', 'accommodates'] > 0.7:
    print("     ‚Üí COLINEALIDAD MODERADA: Considerar multicolinealidad en el modelo")
else:
    print("     ‚Üí Colinealidad aceptable")

print(f"\n   ‚Ä¢ Correlaci√≥n beds-bedrooms: {corr_matrix.loc['beds', 'bedrooms']:.3f}")
print("     ‚Üí Beds es redundante si ya tenemos bedrooms")

# 2. An√°lisis de outliers usando IQR
print("\n\n2. AN√ÅLISIS DE OUTLIERS (m√©todo IQR):")
print("-" * 80)

outlier_analysis = {}
for col in ['price', 'bedrooms', 'accommodates']:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_pct = (len(outliers) / len(df)) * 100
        
        outlier_analysis[col] = {
            'Q1': Q1, 'Q3': Q3, 'IQR': IQR,
            'lower_bound': lower_bound, 'upper_bound': upper_bound,
            'outliers_count': len(outliers), 'outlier_pct': outlier_pct
        }
        
        print(f"\n{col.upper()}:")
        print(f"   Rango normal: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print(f"   Outliers detectados: {len(outliers)} ({outlier_pct:.1f}% del dataset)")
        if col == 'price':
            print(f"   Outliers > ${upper_bound:.2f}")

# 3. Amplitud de variaci√≥n
print("\n\n3. AMPLITUD DE VARIACI√ìN (Coef. Variaci√≥n = std/media):")
print("-" * 80)
for col in ['price', 'bedrooms', 'accommodates']:
    if col in df.columns:
        cv = df[col].std() / df[col].mean()
        print(f"{col}: CV = {cv:.2f} ", end="")
        if cv > 1:
            print("(Alta variabilidad: datos dispersos)")
        else:
            print("(Variabilidad moderada)")
```

# Seccion 4

En esta secci√≥n analizaremos los valores at√≠picos de la variable price. Dado que los precios no siguen una distribuci√≥n normal y presentan un fuerte sesgo, utilizaremos el Rango Intercuart√≠lico en lugar del Z-Score. Este m√©todo es ideal para datos de Airbnb porque utiliza los cuartiles para establecer l√≠mites realistas, permiti√©ndonos identificar registros extremos que podr√≠an distorsionar nuestros promedios y modelos sin depender de la simetr√≠a de la campana de Gauss.


Originalmente, se aplic√≥ un m√©todo de eliminaci√≥n acumulativa basado en el Rango Intercuart√≠lico (IQR) a todas las variables num√©ricas del dataset. Si bien este enfoque es estad√≠sticamente riguroso, gener√≥ un problema cr√≠tico de sobrelimpieza al descartar cualquier registro que presentara un valor at√≠pico en al menos una de las 42 columnas analizadas, lo que redujo dr√°sticamente la muestra de 14913 a solo 3044 registros y elimin√≥ informaci√≥n valiosa de la poblaci√≥n. Para solucionar esta p√©rdida masiva de datos, se propuso una estrategia h√≠brida: realizar la eliminaci√≥n f√≠sica de filas √∫nicamente en la variable objetivo (price), por ser la m√°s sensible al ruido, mientras que para las variables predictoras restantes se opt√≥ por la t√©cnica de Capping o Clip. Este ajuste permite "topar" los valores extremos dentro de los l√≠mites del IQR sin sacrificar el volumen de la muestra, garantizando un equilibrio √≥ptimo entre la calidad estad√≠stica y la representatividad del dataset.

```python
# Funci√≥n para detectar outliers usando IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Configuraci√≥n de columnas y preparaci√≥n
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
final_cols = [
    col for col in num_cols
    if 'id' not in col.lower() 
    and df[col].nunique() > 2 
    and df[col].notnull().any()
]

df_clean = df.copy()

# Eliminar outliers de precio (Variable Cr√≠tica)
low_p, high_p = detect_outliers_iqr(df_clean, 'price')
df_clean = df_clean[(df_clean['price'] >= low_p) & (df_clean['price'] <= high_p)]

# Configuraci√≥n din√°mica de la figura
n_cols_plot = 2
n_rows_plot = math.ceil(len(final_cols) / n_cols_plot)
plt.figure(figsize=(14, 5 * n_rows_plot))

# Bucle de visualizaci√≥n y Capping
for i, col in enumerate(final_cols, 1):
    # Detectar l√≠mites para la columna actual
    low, high = detect_outliers_iqr(df_clean, col)
    n_outliers = ((df_clean[col] < low) | (df_clean[col] > high)).sum() 
    
    # Imputar (Capping)
    df_clean[col] = df_clean[col].clip(lower=low, upper=high)
    
    # Crear Subplot
    plt.subplot(n_rows_plot, n_cols_plot, i)
    sns.boxplot(x=df_clean[col])
    plt.title(f'Boxplot Final (Limpio) - {col}')
    
    print(f"{col}: {n_outliers} outliers procesados. L√≠mites: [{low:.2f}, {high:.2f}]")

plt.tight_layout()
plt.show()

print(f"\nDimensiones finales del dataset: {df_clean.shape}")

df = df_clean.copy()
```

Para esto nos devolvimos a nuestro estudio de correlaci√≥n de variables para poder elegir cual variable nos proporcionaria m√°s informaci√≥n en el modelo. La variable de 'accomodates' es la variable m√°s correlacionada con el precio segun nuestro an√°lisis, por lo que esta variable la compararemos directamente en nuestra regresi√≥n lineal simple. Por otro lado, para nuestro estudio multivariado decidimos escoger las tres variables con correlaciones fuertes. Estas variables fueron 'accomodates', 'bedrooms' y 'number_of_reviews', omitiendo 'beds' ya que puede agregar multicolinearidad y a√±adir o sesgar la informaci√≥n del modelo.   

```python
# Elegimos las variables a evaluar
# Accomodates para el modelo simple
X_simple = df_model[['accommodates']].dropna()
y_simple = y_log[X_simple.index]

# Accomodates, bedrooms y number_of_reviews para el modelo multivariado
features_multi = ['accommodates', 'bedrooms', 'number_of_reviews']

X_multi = df_model[features_multi].dropna()
y_multi = y_log[X_multi.index]
```

Una vez hemos escogido estas variables, empezamos con los modelos. Primero, evaluamos la regresi√≥n lineal con los datos propuestos. Como la variable 'accomodates' es discreta, el grafico nos muestra los valores reales como lineas de puntos en cada unidad en X. Para poder visualizar un poco m√°s los datos, hemos agregado un poco de ruido visual para proporcionar informaci√≥n sobre el volumen de datos en cada valor de X. 

```python
# MODELO DE REGRESION LINEAL SIMPLE

X_train, X_test, y_train, y_test = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)

model_simple = LinearRegression()
model_simple.fit(X_train, y_train)
y_pred = model_simple.predict(X_test)

print(f"R¬≤:          {r2_score(y_test, y_pred):.4f}")
print(f"RMSE:        {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}")
print(f"Coeficiente: {model_simple.coef_[0]:.4f}")
print(f"Intercepto:  {model_simple.intercept_:.4f}")

X_sm = sm.add_constant(X_train)
model_sm = sm.OLS(y_train, X_sm).fit()
print(model_sm.summary())

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

error_factor = np.exp(rmse)
print(f"Factor de error: {error_factor:.4f}")
print(f"Esto significa que el modelo se equivoca por un factor de ~{error_factor:.2f}x")

# Visualizaci√≥n con jitter para mostrar mas detalle (peque√±o ruido visual)
jitter = np.random.uniform(-0.2, 0.2, size=len(X_test))

plt.figure(figsize=(10, 6))
plt.scatter(X_test.values.flatten() + jitter, y_test, alpha=0.2, s=15, label='Datos reales')

x_range = np.linspace(X_test.values.min(), X_test.values.max(), 100).reshape(-1, 1)
plt.plot(x_range, model_simple.predict(x_range), color='red', linewidth=2, label='Recta de regresi√≥n')

plt.xlabel('accommodates')
plt.ylabel('log(price)')
plt.title('Regresi√≥n Lineal Simple')
plt.legend()
plt.show()
```

## Modelo de Regresion Lineal Simple
El modelo de regresi√≥n lineal simple contrasta la variable independiente 'accommodates' con 'log(price)'. El intercepto de 3.899 equivale a un precio base de ~49‚Ç¨ (e^3.899), representando el valor te√≥rico cuando 'accommodates' = 0, un escenario que no ocurre en la pr√°ctica pero que sirve como ancla matem√°tica del modelo. El coeficiente de 'accommodates' es 0.2809, lo que indica que por cada hu√©sped adicional que admite el alojamiento, el precio aumenta aproximadamente un 33% (e^0.2809 - 1 ‚âà 0.325). Esta relaci√≥n multiplicativa es m√°s realista que un incremento fijo en euros, ya que refleja que agregar capacidad tiene un impacto proporcionalmente mayor en alojamientos peque√±os que en grandes.

En cuanto a las m√©tricas de evaluaci√≥n, el R¬≤ de 0.43 indica que el modelo explica el 43% de la variabilidad del precio, lo cual es moderado para una regresi√≥n de una sola variable. El MSE del modelo es de 0.3391 y el RMSE de 0.5823 en escala logar√≠tmica. Para interpretar esto en t√©rminos reales, el RMSE equivale a un factor de error multiplicativo de aproximadamente 1.79x (e^0.5823), lo que significa que para un alojamiento cuyo precio real es 100‚Ç¨, el modelo podr√≠a predecir entre ~56‚Ç¨ y ~179‚Ç¨.

Si bien estas m√©tricas demuestran que 'accommodates' tiene una relaci√≥n relevante con el precio, tambi√©n evidencian las limitaciones de usar una √∫nica variable explicativa. Se hipotetiza que el modelo multivariado, al incorporar 'bedrooms' y 'number_of_reviews', mejorar√° estas m√©tricas por las siguientes razones: 'bedrooms' est√° altamente correlacionada con 'accommodates' pero aporta informaci√≥n adicional sobre la estructura f√≠sica del alojamiento, capturando variabilidad que 'accommodates' sola no recoge. Por su parte, 'number_of_reviews' introduce una dimensi√≥n de demanda y popularidad del alojamiento, factores que influyen en la fijaci√≥n del precio independientemente del tama√±o. En conjunto, se espera un R¬≤ superior a 0.43 y un RMSE menor, reflejando un modelo m√°s preciso y representativo de la realidad del mercado de Airbnb en Barcelona.

```python
# MODELO DE REGRESION LINEAL MULTIVARIADO

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

model_multi = LinearRegression()
model_multi.fit(X_train_m, y_train_m)
y_pred_m = model_multi.predict(X_test_m)

print(f"R¬≤:   {r2_score(y_test_m, y_pred_m):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test_m, y_pred_m)):.4f}")

# Coeficientes de cada variable
coef_df = pd.DataFrame({'feature': features_multi, 'coeficiente': model_multi.coef_})
print("\nCoeficientes:")
print(coef_df)

# Resumen estad√≠stico con p-values (statsmodels)
X_sm = sm.add_constant(X_train_m)
model_sm = sm.OLS(y_train_m, X_sm).fit()
print(model_sm.summary())

mse_m = mean_squared_error(y_test_m, y_pred_m)
error_factor_m = np.exp(np.sqrt(mse_m))
print(f"MSE:          {mse_m:.4f}")
print(f"Factor error: {error_factor_m:.4f}x")
```

## Modelo Multivariado
El modelo de regresi√≥n lineal multivariada incorpora tres variables independientes: 'accommodates', 'bedrooms' y 'number_of_reviews', contrastadas contra 'log(price)'. El intercepto de 3.9062 equivale a un precio base te√≥rico de ~49.7‚Ç¨ (e^3.9062), pr√°cticamente id√©ntico al del modelo simple, lo que indica consistencia entre ambos modelos en su punto de partida.
En cuanto a los coeficientes, 'accommodates' mantiene el mayor peso explicativo con un valor de 0.3064, indicando que por cada hu√©sped adicional el precio aumenta aproximadamente un 35.8% (e^0.3064 - 1), un efecto ligeramente mayor que en el modelo simple debido al ajuste por las otras variables. El coeficiente de 'bedrooms' es negativo (-0.1200), lo que a primera vista puede parecer contraintuitivo, pero se explica por la alta correlaci√≥n entre 'bedrooms' y 'accommodates': al controlar por el n√∫mero de hu√©spedes, tener m√°s habitaciones no necesariamente implica un precio mayor, y puede reflejar alojamientos con habitaciones peque√±as o de menor calidad. Finalmente, 'number_of_reviews' tiene un coeficiente de 0.002565, positivo pero muy peque√±o, sugiriendo que la popularidad del alojamiento tiene un efecto marginal sobre el precio. Cabe destacar que los tres coeficientes tienen p-values de 0.000, confirmando que son estad√≠sticamente significativos.
Respecto al ajuste del modelo, el R¬≤ de 0.48 representa una mejora respecto al modelo simple (0.43), indicando que el modelo multivariado explica el 48% de la variabilidad del precio. El RMSE de 0.5581 es tambi√©n inferior al del modelo simple (0.5823), equivaliendo a un factor de error multiplicativo de ~1.75, una mejora modesta pero consistente.


```python
# Estudio de residuales
residuals = y_test_m - y_pred_m

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(y_pred_m, residuals, alpha=0.3, s=15)
axes[0].axhline(0, color='red', linestyle='--')
axes[0].set_xlabel('Predicciones')
axes[0].set_ylabel('Residuales')
axes[0].set_title('Residuales vs Predicciones')

axes[1].hist(residuals, bins=40, color='steelblue', edgecolor='black')
axes[1].set_title('Distribuci√≥n de Residuales')
plt.tight_layout()
plt.show()
```

## An√°lisis de Residuales
Los gr√°ficos de residuales del modelo multivariado revelan un patr√≥n muy similar al del modelo simple. En el gr√°fico de Residuales vs Predicciones se observa heterocedasticidad: la dispersi√≥n de los errores es mayor para predicciones bajas (~4.0‚Äì4.5) y se reduce progresivamente hacia predicciones m√°s altas, lo que indica que el modelo es menos preciso para alojamientos de precio bajo que para los de precio alto. Este patr√≥n no se corrigi√≥ al agregar 'bedrooms' y 'number_of_reviews', sugiriendo que la heterocedasticidad es una caracter√≠stica estructural de los datos m√°s que una consecuencia de usar pocas variables.
La distribuci√≥n de residuales mantiene una forma aproximadamente normal centrada en 0, lo cual es positivo e indica que el modelo no presenta sesgos sistem√°ticos graves. Sin embargo, se aprecia una leve asimetr√≠a hacia la izquierda, con una cola m√°s pronunciada en valores negativos, lo que refleja que el modelo tiende a sobreestimar el precio en algunos casos de forma m√°s marcada de lo que lo subestima.
En conjunto, los residuales confirman que los supuestos de normalidad se cumplen de manera aceptable, pero la heterocedasticidad presente representa una limitaci√≥n formal del modelo que deber√≠a considerarse al interpretar sus predicciones individuales.
