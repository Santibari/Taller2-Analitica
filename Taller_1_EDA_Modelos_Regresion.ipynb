# Secci√≥n 1:

Durante el proceso de limpieza de datos se tomaron decisiones diferenciadas seg√∫n la importancia de cada variable. Por un lado, se eliminaron las observaciones que presentaban valores faltantes en la variable price, ya que esta representa la variable principal del an√°lisis y es esencial para futuros modelos de regresi√≥n; imputar valores en este caso podr√≠a introducir sesgos y afectar la validez de los resultados. Por otro lado, se realiz√≥ imputaci√≥n en la variable reviews_per_month utilizando la mediana, debido a que es una caracter√≠stica auxiliar y su ausencia no justifica la eliminaci√≥n completa de registros. El uso de la mediana permite conservar la mayor cantidad de datos posibles y reduce el impacto de valores extremos presentes en el dataset. Estas decisiones se basaron en criterios estad√≠sticos y en la relevancia anal√≠tica de cada variable dentro del estudio.

```python
# 1. Revisi√≥n de los datos
df.info()
df.shape
df.dtypes

# Valores faltantes (Top 20)
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)

# Duplicados
df.duplicated().sum()

# Eliminar duplicados (y verificar)
df = df.drop_duplicates()
df.duplicated().sum()


# 2. Correcci√≥n de fechas
df["last_scraped"] = pd.to_datetime(df["last_scraped"], errors="coerce")
df["host_since"]   = pd.to_datetime(df["host_since"], errors="coerce")

# Verificar tipos y NaT
df[["last_scraped", "host_since"]].dtypes
df[["last_scraped", "host_since"]].isna().sum()


# 3. Conversi√≥n de num√©ricas que est√°n como texto

# Porcentajes -> float (ej. "95%" -> 95.0)
for col in ["host_response_rate", "host_acceptance_rate"]:
    if col in df.columns:
        df[col] = (df[col]
                   .astype(str)
                   .str.replace("%", "", regex=False)
                   .replace("nan", np.nan))
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Booleano t/f -> 1/0
if "host_is_superhost" in df.columns:
    df["host_is_superhost"] = (
    df["host_is_superhost"]
    .astype(str)
    .str.lower()
    .map({"t":1, "true":1, "f":0, "false":0})
)

# Price: convertir si viene como texto con $ o comas
if "price" in df.columns and df["price"].dtype == "object":
    df["price"] = (df["price"].astype(str)
                   .str.replace("$", "", regex=False)
                   .str.replace(",", "", regex=False)
                   .replace("nan", np.nan))
    df["price"] = pd.to_numeric(df["price"], errors="coerce")

# (Opcional: evidencia de que cambi√≥)
df[["host_response_rate", "host_acceptance_rate", "host_is_superhost", "price"]].dtypes


# 4. Manejo de valores faltantes

# Eliminar filas si falta variable clave (price)
df = df.dropna(subset=["price"])

# Imputar reviews_per_month
if "reviews_per_month" in df.columns:
    df["reviews_per_month"] = pd.to_numeric(df["reviews_per_month"], errors="coerce")
    df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].median())



# 5. Verificar cambios
df.info()
df.isnull().sum().sort_values(ascending=False).head(20)
df.describe()
df.shape
```

```python
# Importar librer√≠as necesarias para visualizaci√≥n
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ===== 2.1 ESTAD√çSTICAS DESCRIPTIVAS =====
print("="*80)
print("ESTAD√çSTICAS DESCRIPTIVAS - VARIABLES NUM√âRICAS")
print("="*80)

# Variables clave para an√°lisis
variables_key = ['price', 'bedrooms', 'beds', 'accommodates', 'number_of_reviews', 
                 'review_scores_rating', 'reviews_per_month']

# Filtrar variables que existen en el dataset
variables_key = [col for col in variables_key if col in df.columns]

# Estad√≠sticas descriptivas
print("\nEstad√≠sticas Descriptivas de Variables Clave:")
print(df[variables_key].describe().round(2))

# Estad√≠sticas adicionales
print("\n\nEstad√≠sticas Adicionales:")
print(f"Rango de precio: ${df['price'].min():.2f} - ${df['price'].max():.2f}")
print(f"Desviaci√≥n est√°ndar del precio: ${df['price'].std():.2f}")
print(f"Coeficiente de variaci√≥n (CV) del precio: {(df['price'].std()/df['price'].mean()):.2f}")
print(f"\nAsimetr√≠a (skewness) de price: {df['price'].skew():.2f}")
print(f"Curtosis de price: {df['price'].kurtosis():.2f}")

# Proporciones por tipo de propiedad si existe la columna
if 'room_type' in df.columns:
    print("\n\nDistribuci√≥n por Tipo de Habitaci√≥n:")
    print(df['room_type'].value_counts())
    print("\nPrecio promedio por tipo de habitaci√≥n:")
    print(df.groupby('room_type')['price'].agg(['mean', 'median', 'std', 'count']).round(2))

# Proporciones por superhosts
if 'host_is_superhost' in df.columns:
    print("\n\nSuperhost vs No-Superhost:")
    print(df['host_is_superhost'].value_counts())
    print("\nPrecio promedio por superhost status:")
    print(df.groupby('host_is_superhost')['price'].agg(['mean', 'median', 'std', 'count']).round(2))
```

```python
# ===== AN√ÅLISIS COMPLEMENTARIO: COLINEALIDAD Y OUTLIERS =====

print("\n" + "="*80)
print("AN√ÅLISIS T√âCNICO COMPLEMENTARIO")
print("="*80)

# 1. Matriz de correlaci√≥n entre variables num√©ricas clave
print("\n1. MATRIZ DE CORRELACI√ìN (para detectar colinealidad):")
print("-" * 80)
numeric_cols = ['price', 'bedrooms', 'beds', 'accommodates', 'number_of_reviews']
numeric_cols = [col for col in numeric_cols if col in df.columns]
corr_matrix = df[numeric_cols].corr()
print(corr_matrix.round(3))

# mapa de calor de correlaciones
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor - Correlaciones')
plt.show()

print("\n‚ö†Ô∏è  AN√ÅLISIS DE COLINEALIDAD:")
print(f"   ‚Ä¢ Correlaci√≥n bedrooms-accommodates: {corr_matrix.loc['bedrooms', 'accommodates']:.3f}")
if corr_matrix.loc['bedrooms', 'accommodates'] > 0.8:
    print("     ‚Üí ALTA COLINEALIDAD: Usar solo accommodates en el modelo")
elif corr_matrix.loc['bedrooms', 'accommodates'] > 0.7:
    print("     ‚Üí COLINEALIDAD MODERADA: Considerar multicolinealidad en el modelo")
else:
    print("     ‚Üí Colinealidad aceptable")

print(f"\n   ‚Ä¢ Correlaci√≥n beds-bedrooms: {corr_matrix.loc['beds', 'bedrooms']:.3f}")
print("     ‚Üí Beds es redundante si ya tenemos bedrooms")

# 2. An√°lisis de outliers usando IQR
print("\n\n2. AN√ÅLISIS DE OUTLIERS (m√©todo IQR):")
print("-" * 80)

outlier_analysis = {}
for col in ['price', 'bedrooms', 'accommodates']:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_pct = (len(outliers) / len(df)) * 100
        
        outlier_analysis[col] = {
            'Q1': Q1, 'Q3': Q3, 'IQR': IQR,
            'lower_bound': lower_bound, 'upper_bound': upper_bound,
            'outliers_count': len(outliers), 'outlier_pct': outlier_pct
        }
        
        print(f"\n{col.upper()}:")
        print(f"   Rango normal: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print(f"   Outliers detectados: {len(outliers)} ({outlier_pct:.1f}% del dataset)")
        if col == 'price':
            print(f"   Outliers > ${upper_bound:.2f}")

# 3. Amplitud de variaci√≥n
print("\n\n3. AMPLITUD DE VARIACI√ìN (Coef. Variaci√≥n = std/media):")
print("-" * 80)
for col in ['price', 'bedrooms', 'accommodates']:
    if col in df.columns:
        cv = df[col].std() / df[col].mean()
        print(f"{col}: CV = {cv:.2f} ", end="")
        if cv > 1:
            print("(Alta variabilidad: datos dispersos)")
        else:
            print("(Variabilidad moderada)")
```
